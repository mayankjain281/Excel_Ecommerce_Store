 If there are multiple sheets in the excel workbook, the command will import data of the first sheet. To make a data frame with all the sheets in the workbook, the easiest method is to create different data frames separately and then concatenate them.

```python
file_name = '/content/EXCEL PORT RO.xlsx'
df = pd.read_excel(file_name)
```

```python
df
# first sheet of workbook (sheet_name=0)
```

```python
# Read Excel by Sheet Number
sheet1 = pd.read_excel(file_name, sheet_name=7)
sheet1.head(3)
```

```python
# Read Excel by Sheet Name
sheet1 = pd.read_excel(file_name, sheet_name='Sheet1')
sheet1.head(3)
```

```python
# Read Excel by Sheet Name
sheet1 = pd.read_excel(file_name, sheet_name=1, header=2)
sheet1.head(3)
```

```python
sheets = pd.read_excel(file_name, sheet_name=[0,1], header=2)
Sheet1_df = sheets.get(0)
Sheet2_df = sheets.get(1)
```
